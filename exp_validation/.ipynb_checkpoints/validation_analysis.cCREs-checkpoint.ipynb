```python
#load in original bed from prediction
original_bed = pybedtools.BedTool("/gpfs/ysm/scratch60/gerstein/zc264/ChromVar/enhancer-prediction/encode/datasets/validation/Sutton/H1.17k.enhancer.hg19.bed").sort()

#load in refined bed from prediction
refined_bed = pybedtools.BedTool("/gpfs/ysm/scratch60/gerstein/zc264/ChromVar/enhancer-prediction/encode/exp_validation/H1.tested_17k_sites.50.4kb.10bp_bins.Grad-CAM.filtered.hg19.lifted.bed").sort()

#load in bed refined by grad cam
ccre_bed = pybedtools.BedTool("/gpfs/ysm/scratch60/gerstein/zc264/ChromVar/enhancer-prediction/encode/datasets/H1/H1.cCRE.hg19.lifted.bed").sort()
```

```python
gnomadrare="/gpfs/slayman/pi/gerstein/yy555/Project/RareDAF_redo/New_data_set_9/GetLocal_2/data/gnomad.rare.noncoding.bw"
gnomadcomm="/gpfs/slayman/pi/gerstein/yy555/Project/RareDAF_redo/New_data_set_9/GetLocal_2/data/gnomad.comm.noncoding.bw"
pcawgrare="/gpfs/slayman/pi/gerstein/yy555/Project/RareDAF_redo/Variant_rare_comm_3/variant_bigwig/pcawg.noncoding.rare.0.0004.bw"
pcawgcomm="/gpfs/slayman/pi/gerstein/yy555/Project/RareDAF_redo/Variant_rare_comm_3/variant_bigwig/pcawg.noncoding.comm.0.01.bw"

rare_daf = [gnomadrare, gnomadcomm, pcawgrare, pcawgcomm]
refined_daf = []
original_daf = []
ccre_daf = []

for r in rare_daf:
    rare_daf_bw = pyBigWig.open(r)
    
    
    refined_raredaf = []
    refined_bed = pybedtools.BedTool("/gpfs/ysm/scratch60/gerstein/zc264/ChromVar/enhancer-prediction/encode/exp_validation/H1.tested_17k_sites.50.4kb.10bp_bins.Grad-CAM.filtered.hg19.lifted.bed").sort()
    for x in refined_bed:
        try:
            results = np.nansum(rare_daf_bw.values(x.chrom, x.start, x.stop))
        except:
            continue
        refined_raredaf.append(results)
    print("refined done")
    
    original_raredaf = []
    original_bed = pybedtools.BedTool("/gpfs/ysm/scratch60/gerstein/zc264/ChromVar/enhancer-prediction/encode/datasets/validation/Sutton/H1.17k.enhancer.hg19.bed").sort()
    for x in original_bed:
        try:
            results = np.nansum(rare_daf_bw.values(x.chrom, x.start, x.stop))
        except:
            continue
        original_raredaf.append(results)
    print("original done")
    
    ccre_raredaf = []
    ccre_bed = pybedtools.BedTool("/gpfs/ysm/scratch60/gerstein/zc264/ChromVar/enhancer-prediction/encode/datasets/H1/H1.cCRE.hg19.lifted.bed").sort()
    for x in ccre_bed:
        try:
            results = np.nansum(rare_daf_bw.values(x.chrom, x.start, x.stop))
        except:
            continue
        ccre_raredaf.append(results)
    print("ccre done")
    
    
    refined_signal = np.sum(np.nan_to_num(np.array(refined_raredaf)))
    original_signal = np.sum(np.nan_to_num(np.array(original_raredaf)))
    ccre_signal = np.sum(np.nan_to_num(np.array(ccre_raredaf)))
    
    refined_daf.append(refined_signal)
    original_daf.append(original_signal)
    ccre_daf.append(ccre_signal)
    print(refined_signal, original_signal, ccre_signal)
```

```python
gnomad_refined = refined_daf[0] / (refined_daf[0] + refined_daf[1])
gnomad_original = original_daf[0] / (original_daf[0] + original_daf[1])
gnomad_ccre = ccre_daf[0] / (ccre_daf[0] + ccre_daf[1])
print(gnomad_refined, gnomad_original, gnomad_ccre)

pcawg_refined = refined_daf[2] / (refined_daf[2] + refined_daf[3])
pcawg_original = original_daf[2] / (original_daf[2] + original_daf[3])
pcawg_ccre = ccre_daf[2] / (ccre_daf[2] + ccre_daf[3])
print(pcawg_refined, pcawg_original, pcawg_ccre)
```

```python
from scipy import stats
gnomad_p = -np.log10(stats.binom_test(original_daf[0], n=(original_daf[0] + original_daf[1]), p=gnomad_ccre, alternative='greater'))
pcawg_p = -np.log10(stats.binom_test(original_daf[2], n=(original_daf[2] + original_daf[3]), p=pcawg_ccre, alternative='greater'))
print(gnomad_p, pcawg_p)
```

```python
refined_results = glob.glob("/gpfs/ysm/scratch60/gerstein/zc264/ChromVar/enhancer-prediction/encode/exp_validation/H1_output/*.17k.enhancer.hg19.bed.results")
original_results = glob.glob("/gpfs/ysm/scratch60/gerstein/zc264/ChromVar/enhancer-prediction/encode/exp_validation/H1_output/*.tested_17k_sites.50.4kb.10bp_bins.Grad-CAM.filtered.results")
ccre_results = glob.glob("/gpfs/ysm/scratch60/gerstein/zc264/ChromVar/enhancer-prediction/encode/exp_validation/H1_output/*H1.cCRE.results")
```

```python
from scipy.stats import norm

refined_ldsc = []
for i in refined_results:
    value = pd.read_csv(i, sep="\t").iloc[0, 9]
    refined_ldsc.append(-np.log10(norm.sf(value)))

original_ldsc = []
for i in original_results:
    value = pd.read_csv(i, sep="\t").iloc[0, 9]
    original_ldsc.append(-np.log10(norm.sf(value)))
    
ccre_ldsc = []
for i in ccre_results:
    value = pd.read_csv(i, sep="\t").iloc[0, 9]
    ccre_ldsc.append(-np.log10(norm.sf(value)))
```

```python
result_names = [i.replace("/gpfs/ysm/scratch60/gerstein/zc264/ChromVar/enhancer-prediction/encode/exp_validation/H1_output/", "") for i in original_results]
result_names = [i.replace(".H1.tested_17k_sites.50.4kb.10bp_bins.Grad-CAM.filtered.results", "") for i in result_names]
result_names = [i.replace("_", " ") for i in result_names]
result_names[result_names.index("DS")] = "Depression"
result_names[result_names.index("BMI1")] = "BMI"
result_names[result_names.index("Height1")] = "Height"
result_names[result_names.index("Years of Education1")] = "Years of Primary Education"
result_names[result_names.index("Years of Education2")] = "Years of Secondary Education"
result_names[result_names.index("SWB")] = "State of Well-being"
result_names
```

```python
matplotlib.rcParams.update({'font.size': 22})


labels = result_names

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(12,6))
rects1 = ax.bar(x - width/2, refined_ldsc, width, label='Refined', color = ["#1F78B4"])
rects2 = ax.bar(x + width/2, ccre_ldsc, width, label='ENCODE cCRE', color = ["#FF7F50"])

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('-log(p-val)')
ax.set_title('LDSC Enrichment Analysis')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()
plt.axhline(y=1.301, color="r", linestyle='--')
plt.xticks(rotation=90)
plt.show()
```
